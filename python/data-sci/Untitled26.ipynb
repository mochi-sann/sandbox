## 商品名の表記揺れの統一

### Subtask:
商品名 列の表記揺れを修正します。


**Reasoning**:
Identify and unify the variations in the '商品名' column to address the data cleaning subtask. Then, display the unique values in the '商品名' column to confirm the changes.



```python
print("Original unique values in '商品名':")
display(uriage_data['商品名'].unique())

uriage_data['商品名'] = uriage_data['商品名'].replace('はさみ', 'ハサミ')

print("\nUnique values in '商品名' after cleaning:")
display(uriage_data['商品名'].unique())
```

## 欠損値の処理

### Subtask:
欠損値が存在する列に対して適切な処理を行います。


**Reasoning**:
Fill missing values in '曜日', '顧客の性別', and '顧客の年齢' columns with appropriate values.



```python
uriage_data['曜日'].fillna('不明', inplace=True)
uriage_data['顧客の性別'].fillna('不明', inplace=True)
uriage_data['顧客の年齢'].fillna(uriage_data['顧客の年齢'].mean(), inplace=True)
display(uriage_data.isnull().sum())
```

### Subtask:

Convert the '顧客の年齢' column to integer type and display the dataframe information to confirm the data types.



```python
uriage_data['顧客の年齢'] = uriage_data['顧客の年齢'].astype(int)
uriage_data.info()
```
